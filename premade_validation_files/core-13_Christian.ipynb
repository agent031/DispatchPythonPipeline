```python
def rescale(sn,center=[0,0,0],to='au'):
    for p in sn.patches:
        e=p.extent
        e[0][0:2] -= center[1]
        e[0][2:4] -= center[2]
        e[1][0:2] -= center[0]
        e[1][2:4] -= center[2]
        e[2][0:2] -= center[0]
        e[2][2:4] -= center[1]
        if to=='au':
            e=e*sn.scaling.l/sn.cgs.au
        elif to=='pc:
            e=e*sn.scaling.l/sn.cgs.pc
```
